```python
# Histograms & Boxplots

import matplotlib.pyplot as plt

df.hist(figsize=(10,8))
plt.show()
```

```python
#Boxplot (outliers)

import seaborn as sns

for col in df.select_dtypes(include=['int64','float64']).columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Pairplot

sns.pairplot(df)
plt.show()

```

```python
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

# Identifying Patterns and Trends

Age vs Survived → younger survived more

Fare vs Survived → higher fare → higher survival

Pclass vs Fare → higher class → higher fare

 **Feature-Level Inferences**

Age: Most passengers are between 20–40 years. Some missing values were present and handled. Distribution is slightly skewed.

Fare: Highly right-skewed with several extreme values (outliers). Indicates that a few passengers paid significantly higher fares.

Pclass: Passengers in higher class (1st class) tend to have higher fares and better survival chances.

Sex: Female passengers had a higher survival rate compared to males.
SibSp & Parch: Most passengers were traveling alone (values mostly 0), indicating limited family travel.

Correlation: Fare and Pclass show strong relationship (higher class - higher fare).

Outliers: Significant outliers observed in the Fare column.
